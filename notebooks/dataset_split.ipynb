2. Duplicates (exact copies)

```python
duplicates_records = df.sample(n=duplicates_size, random_state=42).copy()
duplicates_records.count()
```

```python
scd1_records = df.sample(n=scd1_size).copy()
# Keep existing order_id, but update data (SCD1)
scd1_records['customer_name'] = scd1_records['customer_name'] + ' (Updated)'
scd1_records['sales'] = scd1_records['sales'] * 1.1
```

```python
last_row_id = df['row_id'].max()
```

```python
scd1_records['valid_from'] = pd.to_datetime(scd1_records['order_date']).dt.strftime('%Y-%m-%d')
scd1_records['valid_to'] = ''
scd1_records['previous_region'] = ''
scd1_records['is_new_version'] = False  # This is an UPDATE, not a new version
scd1_records['scd2_attribute_changed'] = ''

# Create new row_id for SCD1 (to avoid conflict with initial)
scd1_records['row_id'] = range(last_row_id + new_records_size + 1, 
                               last_row_id + new_records_size + len(scd1_records) + 1)
```

```python
scd1_records.count()
```

```python
# 3. SCD TYPE 2 CHANGES (Versioning)

# Select customers for SCD2 changes (those who already have orders)
customer_counts = df['customer_id'].value_counts()
customers_with_history = customer_counts[customer_counts > 1].index.tolist()[:scd2_size]

scd2_records_list = []

# Date for the new version (e.g., one year after the last order)
base_new_date = datetime(2025, 1, 1)

for i, customer_id in enumerate(customers_with_history[:scd2_size]):
    # Get the last record for this customer
    customer_data = df[df['customer_id'] == customer_id].iloc[-1].copy()
    
    # Save the OLD region
    old_region = customer_data['region']
    
    # Select a NEW region (must be different from the old one)
    available_regions = ['South', 'West', 'Central', 'East']
    if old_region in available_regions:
        available_regions.remove(old_region)
    new_region = random.choice(available_regions) if available_regions else 'Central'
    
    # Create a new version of the record
    new_version = customer_data.copy()
    
    # Change SCD2 attribute (region)
    new_version['region'] = new_region
    
    # Save the old region in a separate field
    new_version['previous_region'] = old_region
    
    # Create a new order_id for the new version
    original_order_id = new_version['order_id']
    new_version['order_id'] = f"SCD2-{original_order_id}-V2"
    
    # Set new dates (later than old ones)
    old_date = pd.to_datetime(new_version['order_date'])
    new_date = base_new_date + timedelta(days=i)
    
    new_version['order_date'] = new_date.strftime('%m/%d/%Y')
    new_version['ship_date'] = (new_date + timedelta(days=random.randint(1, 5))).strftime('%m/%d/%Y')
    
    # Change some data for realism
    new_version['sales'] = (new_version['sales'] * random.uniform(0.9, 1.2)).round(2)
    new_version['quantity'] = new_version['quantity'] + random.randint(-1, 2)
    
    # new row_id
    start_id = last_row_id + new_records_size + len(scd1_records) + i + 1
    new_version['row_id'] = start_id
    
    new_version['valid_from'] = new_date.strftime('%Y-%m-%d')  # When the new version started
    new_version['valid_to'] = ''  # Empty = current version
    new_version['is_new_version'] = True  # This is a NEW version
    new_version['scd2_attribute_changed'] = 'region'  # Which attribute changed
    
    scd2_records_list.append(new_version)

scd2_records = pd.DataFrame(scd2_records_list)

print(f"SCD2 records created: {len(scd2_records)}")
print(f"Example: Customer {scd2_records.iloc[0]['customer_id']}")
print(f"  Old region: {scd2_records.iloc[0]['previous_region']}")
print(f"  New region: {scd2_records.iloc[0]['region']}")
```

```python
# For new records
new_records['valid_from'] = pd.to_datetime(new_records['order_date']).dt.strftime('%Y-%m-%d')
new_records['valid_to'] = ''
new_records['previous_region'] = ''
new_records['is_new_version'] = True
new_records['scd2_attribute_changed'] = ''

# For duplicates
duplicates_records['valid_from'] = pd.to_datetime(duplicates_records['order_date']).dt.strftime('%Y-%m-%d')
duplicates_records['valid_to'] = ''
duplicates_records['previous_region'] = ''
duplicates_records['is_new_version'] = False
duplicates_records['scd2_attribute_changed'] = ''
```

```python
version_columns = ['valid_from', 'valid_to', 'previous_region', 'is_new_version', 'scd2_attribute_changed']

for df_name, temp_df in [('new_records', new_records), 
                         ('duplicates_records', duplicates_records),
                         ('scd1_records', scd1_records),
                         ('scd2_records', scd2_records)]:
    for col in version_columns:
        if col not in temp_df.columns:
            print(f"Adding column {col} to {df_name}")
            if col == 'is_new_version':
                temp_df[col] = False
            elif col == 'scd2_attribute_changed':
                temp_df[col] = ''
            else:
                temp_df[col] = ''
```

```python
secondary_df = pd.concat([new_records, duplicates_records, scd1_records, scd2_records], ignore_index=True)
```

```python
# Create full list of columns
base_columns = df.columns.tolist()
version_columns = ['valid_from', 'valid_to', 'previous_region', 'is_new_version', 'scd2_attribute_changed']
all_columns = base_columns + version_columns

# Ensure secondary_df has all columns
for col in all_columns:
    if col not in secondary_df.columns:
        secondary_df[col] = ''

# Reorder columns
secondary_df = secondary_df[all_columns]
```

```python
# Fix quantity = 0 to 1 (minimum value)
def fix_quantity(df):
    df['quantity'] = df['quantity'].apply(lambda x: 1 if x == 0 else x)
    return df

# Apply to all parts
new_records = fix_quantity(new_records)
scd1_records = fix_quantity(scd1_records)
scd2_records = fix_quantity(scd2_records)
duplicates_records = fix_quantity(duplicates_records)

# Or simply for secondary_df after creation:
secondary_df['quantity'] = secondary_df['quantity'].apply(lambda x: 1 if x == 0 else x)

# Check
print(f"Records with quantity = 0: {(secondary_df['quantity'] == 0).sum()}")
```

```python
# 1. List of date columns
date_cols = ['order_date', 'ship_date', 'valid_from']

# 2. Convert everything to YYYY-MM-DD format (2025-01-31)
for col in date_cols:
    # First convert to datetime objects (pandas understands different formats)
    secondary_df[col] = pd.to_datetime(secondary_df[col], errors='coerce')
    # Convert back to strict YYYY-MM-DD string format
    secondary_df[col] = secondary_df[col].dt.strftime('%Y-%m-%d')

# Fix empty valid_to (replace NaT/NaN with NULL or empty string for CSV)
secondary_df['valid_to'] = pd.to_datetime(secondary_df['valid_to'], errors='coerce')
# For CSV, it's better to leave it empty if there is no date
secondary_df['valid_to'] = secondary_df['valid_to'].dt.strftime('%Y-%m-%d').fillna('')

# 3. Ensure quantity is not 0
secondary_df['quantity'] = secondary_df['quantity'].apply(lambda x: 1 if x <= 0 else x)
```

```python
# Engine
engine = get_db_connection()

# 2. dtype_map with SA types (fix error — objects, not strings!)
dtype_map = {
    'row_id': sa_types.INTEGER(),  # SA object for Int64 → INTEGER
    'order_id': sa_types.String(50),
    'order_date': sa_types.String(20),
    'ship_date': sa_types.String(20),
    'ship_mode': sa_types.String(50),
    'customer_id': sa_types.String(20),
    'customer_name': sa_types.String(100),
    'segment': sa_types.String(50),
    'country': sa_types.String(50),
    'city': sa_types.String(50),
    'state': sa_types.String(50),
    'postal_code': sa_types.INTEGER(),
    'region': sa_types.String(50),
    'product_id': sa_types.String(20),
    'category': sa_types.String(50),
    'sub_category': sa_types.String(50),
    'product_name': sa_types.String(200),
    'sales': sa_types.DECIMAL(precision=10, scale=2),
    'quantity': sa_types.INTEGER(),
    'discount': sa_types.DECIMAL(precision=5, scale=2),
    'profit': sa_types.DECIMAL(precision=10, scale=2),
    'valid_from': sa_types.String(20),
    'valid_to': sa_types.String(20),
    'previous_region': sa_types.String(50),
    'is_new_version': sa_types.Boolean(),
    'scd2_attribute_changed': sa_types.String(50)
}
```

```python
# 3. Load (now without errors)
df.to_sql(
    name='raw_orders',
    con=engine,
    schema='stage',
    if_exists='replace',
    index=False,
    method='multi',
    chunksize=1000,
    dtype=dtype_map  # SA objects — fix!
)
```
